```python
# Install necessary libraries (uncomment if not already installed)
# !pip install pandas matplotlib statsmodels

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the dataset (replace the path with your actual file path in Colab)
df = pd.read_csv('/content/Electric_Production.csv', parse_dates=['DATE'], index_col='DATE')

# Display the first few rows of the dataset to check column names
print(df.head())

# Use the electric production data for decomposition
time_series = df['IPG2211A2N']  # Adjust with your actual column name if necessary

# Decompose the time series into trend, seasonal, and residual components
decomposition = seasonal_decompose(time_series, model='multiplicative', period=12)

# Plot the decomposition
plt.figure(figsize=(12, 8))

# Plot the observed data
plt.subplot(411)
plt.plot(time_series, label='Observed')
plt.title('Observed Time Series')
plt.legend(loc='upper left')

# Plot the trend component
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend', color='orange')
plt.title('Trend')
plt.legend(loc='upper left')

# Plot the seasonal component
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonality', color='green')
plt.title('Seasonality')
plt.legend(loc='upper left')

# Plot the residual component
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual', color='red')
plt.title('Residual')
plt.legend(loc='upper left')

# Tighten the layout and show the plot
plt.tight_layout()
plt.show()
```
